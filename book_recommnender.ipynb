# popularity based 

```python
rating_with_name=rating.merge(books,on="ISBN")
```

```python
rating_with_name
```

```python
df=rating_with_name.groupby("Book-Title").count()["Book-Rating"].reset_index()
df.rename(columns={"Book-Rating":"num_rating"},inplace=True)
df.shape
```

```python
df1=rating_with_name.groupby("Book-Title").mean(numeric_only=True)["Book-Rating"].reset_index()
df1.rename(columns={"Book-Rating":"avg_rating"},inplace=True)
df1.shape
```

```python
popular_df=df.merge(df1,on="Book-Title")
```

```python
popular_df.head()
```

```python
popular_df=popular_df[popular_df["num_rating"]>250].sort_values("avg_rating",ascending=False)
```

```python
popular_df
```

```python
popular_df=popular_df.merge(books,on="Book-Title").drop_duplicates("Book-Title")[["Book-Title","Book-Author","Image-URL-S","num_rating","avg_rating"]]
```

```python
popular_df.head()
```

```python
x=rating_with_name.groupby("User-ID").count()["Book-Rating"]>200
good_user=x[x].index
```

```python
filtered=rating_with_name[rating_with_name["User-ID"].isin(good_user)]
```

```python
filtered
```

```python
y=filtered.groupby("Book-Title").count()["Book-Rating"]>=50
famous_book=y[y].index
```

```python
final_rating=filtered[filtered["Book-Title"].isin(famous_book)]
```

```python
final_rating.drop_duplicates()
```

```python
pd=final_rating.pivot_table(index="Book-Title",columns="User-ID",values="Book-Rating")
```

```python
pd.fillna(0,inplace=True)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cosine_similarity(pd).shape
```

```python
score=cosine_similarity(pd)
```

```python
def recommend(book_name):
    index=np.where(pd.index==book_name)[0][0]
    similar_items=sorted(list(enumerate(score[index])),key=lambda x:x[1],reverse=True)[1:6]
    
    for i in similar_items:
        print(pd.index[i[0]])
         
```

```python
recommend("The Notebook")
```
